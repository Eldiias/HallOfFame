```python
def get_last_update(jj):
    for k, v in reversed(list(jj.items())):
        if int(v)>0:
            return k
            break
        if k==list(jj.keys())[0]:
            return None
```

```python
def get_list_of_data(profile):
    j={i.get('data-date'):i.get('data-count') for i in BeautifulSoup(r.get(f'https://github.com/{profile}', headers=headers).content).select('svg>g rect')}
    return get_last_update(j)
```

```python
with open("README.md", "r", encoding="utf-8") as input_file:
    text = input_file.read()
```

```python
import re
```

```python
users=re.findall('https://github.com/(.*?)[ )]', text)
```

```python
jj={i:get_list_of_data(i) for i in users}
```

```python
jj
```

```python
import pandas as pd
```

```python
df=pd.DataFrame.from_dict(jj, orient='index', columns=['Last_Update'])
```

```python
df.dropna(inplace=True)
```

```python
df.Last_Update=pd.to_datetime(df.Last_Update)
```

```python
from datetime import datetime
```

```python
df['Difference']=(datetime.now()-df.Last_Update).dt.days.apply(lambda x: None if x>180 else x)
```

```python
df.dropna(inplace=True)
```
